```python
import turtle

t=turtle.Turtle()
t.shape()
t.hideturtle()

t.penup()
t.goto(-200,-150)
t.pendown()

#vẽ màu đen
t.color('black')
t.fillcolor('black')
t.begin_fill()
t.forward(400)
t.left(90)
t.forward(100)
t.left(90)
t.forward(400)
t.left(90)
t.forward(100)
t.left(90)
t.end_fill()

t.penup()
t.left(90)
t.forward(100)
t.right(90)
t.pendown()

t.color('white')
t.fillcolor('white')
t.begin_fill()
t.forward(400)
t.left(90)
t.forward(100)
t.left(90)
t.forward(400)
t.left(90)
t.forward(100)
t.left(90)
t.end_fill()


t.penup()
t.left(90)
t.forward(100)
t.right(90)
t.pendown()

t.color('red')
t.fillcolor('red')
t.begin_fill()
t.forward(400)
t.left(90)
t.forward(100)
t.left(90)
t.forward(400)
t.left(90)
t.forward(100)
t.left(90)
t.end_fill()

t.penup()
t.right(37)
t.forward(83.2)
t.left(37)
t.pendown()

t.color('green')
t.fillcolor('green')
t.begin_fill()
t.forward(67)
t.right(144)
t.forward(67)
t.right(144)
t.forward(67)
t.right(144)
t.forward(67)
t.right(144)
t.forward(67)
t.end_fill()

t.penup()
t.goto(200,50)
t.right(144)
t.right(143)
t.forward(83.2)
t.right(37)

t.color('green')
t.fillcolor('green')
t.begin_fill()
t.forward(67)
t.left(144)
t.forward(67)
t.left(144)
t.forward(67)
t.left(144)
t.forward(67)
t.left(144)
t.forward(67)
t.left(144)
t.end_fill()

turtle.done()
```
