```python
import csv
with open('D:\Python cơ bản\Python_Fundamental\Object\expense_management.csv','w',encoding='utf-8', newline='') as csv_file:
    fieldnames = ['Expense_id','Description', 'Expense','Date']
    writer = csv.writer(csv_file)
    writer.writerow(fieldnames)
class mucchitieu:
    def __init__(self, id, description, expense, date):
        self.expense_id = id
        self.description = description
        self.expense = expense
        self.date = date 
    def info(self):
        row = self.expense_id.ljust(15, ' ')+ self.description.ljust(60, ' ') + self.expense.ljust(25, ' ') + self.date.ljust(20,' ')
        print(row)
class listchitieu:
    def __init__(self):
        self.list_expense = []
    def Loadfile(self):
        with open('D:\Python cơ bản\Python_Fundamental\Object\expense_management.csv',encoding='utf-8', newline='') as csv_file:
            #Chuyển thành list (nhưng list này chưa dùng đc, vì chưa chuyển thành list các class con)
            list_expense = list(csv.reader(csv_file,delimiter="," ))
            for i in list_expense:
                expense = mucchitieu(i[0],i[1],i[2],i[3]) #chuyển các list nhỏ thành 1 class
                self.list_expense.append(expense)
    def add(self, expense):
        self.list_expense.append(expense)
    def delete(self, expense):
        self.list_expense.remove(expense)
    def showinfo(self):
        print(''.ljust(90, '='))
        for i in self.list_expense:
            i.info()

    def save_file(self):
        confirm = input(f'Enter yes if you want to save- no if unsave: ')
        if confirm.lower().strip() == 'yes':
            with open('D:\Python cơ bản\Python_Fundamental\Object\expense_management.csv','w',encoding='utf-8', newline='') as csv_file:
                write = csv.writer(csv_file)
                for i in self.list_expense:
                    write_row = [i.expense_id, i.description, i.expense, i.date]
                    write.writerow(write_row)
class program:
    def __init__(self):
        self.list = listchitieu()
        self.list.Loadfile()
    def run(self):
        while 1:
            print('============ QUẢN LÝ DANH MUC CHI TIÊU ============')
            print('=== 1. Hiển thị danh mục                ===')
            print('=== 2. Thêm danh mục                    ===')
            print('=== 3. Xóa danh mục                     ===')
            print('=== 4. Lưu thay đổi                     ===')
            print('=== 5. Kết thúc chương trình            ===')
            choice = input('Nhập lựa chọn: ')
            if choice =='1':
                self.list.showinfo()
                option = input('Enter if you want to continue: ')
        
            elif choice == '2':
                expense_id = input('Nhập mã hạng mục: ')
                description = input('Nhập thông tin danh mục chi tiêu: ')
                expense = input('Nhập mức chi tiêu: ')
                date = input('Nhập ngày chi tiêu: ')
                expense = mucchitieu(expense_id, description, expense, date)
                self.list.add(expense)
                self.list.showinfo()
                option = input('Enter if you want to continue: ')
                
            elif choice =='3':
                expense_id = input('Nhập mã hạng mục muốn xóa: ')
                for i in self.list.list_expense:
                    if i.expense_id == expense_id:
                        confirm = input(f'Enter yes if you want to delete {expense_id}')
                        if confirm =='yes':
                            self.list.delete(i)
                            print('Delete completed')
                            break
                        else:
                            option = input('Enter if you want to continue: ')
                    else:
                        print(f'Mã {expense_id} không tồn tại')
                option = input('Enter if you want to continue: ')
                
            elif choice == '4':
                self.list.save_file()
                option = input('Enter if you want to continue: ')
                
            else:
                confirmed = input('Enter True if you want to end: ')
                if confirmed.lower() == 'true':
                    break 
                else: 
                    option = input('Enter if you want to continue: ')
```

```python
run = program()
run.run()
```
